```python
from glob import glob
import os

from imagedaemon.utils.image import Image
from imagedaemon.cameras.winter.winter_image import WinterImage
from imagedaemon.utils.paths import RAW_DATA_DIR

# try to import a qcmos image
image_dir = os.path.join(RAW_DATA_DIR, "qcmos", "science")
image_files = glob(os.path.join(image_dir, "*.fits"))

# get the first image that has 8s in the filename
image_file = None
for image in image_files:
    if "8s" in os.path.basename(image):
        image_file = image
        break
if image_file is None:
    raise RuntimeError("No qcmos image with 8s in the filename found.")

# create an Image object
image = Image(image_file)

image.plot_image(title = "qCMOS 8 s Image", norm="zscale", cmap = "gray")

# load a winter image
image_dir = os.path.join(RAW_DATA_DIR, "winter", "science", "120s")
image_files = sorted(glob(os.path.join(image_dir, "*.fits")))
# get the first image from the list
image_file = image_files[0]
image = WinterImage(image_file)
```
